```python
import requests
import re 
class Webcrawler:
  def __init__(self):
        self.dependent_url =[]
        
  def crawler(self,start_url):
    queue = [start_url]
    self.dependent_url.append(start_url)
    
    while queue:
        
        listed_website = queue.pop(0)
        print(listed_website)
        
        get_html = self.read_raw_url(listed_website)
        
        for url in self.get_http(get_html):
            if url not in self.dependent_url:
                self.dependent_url.append(url)
                queue.append(url)
        
  def get_http(self, website):
    return re.findall("https?://(?:[-\w.]|(?:%[\da-fA-F]{2}))+", website)

  def read_raw_url(self, website):
    
    html = ''
    
    try:
        html = requests.get(website).text

    except Exception as e :
        print(e)
        pass
    return html
    
if __name__=='__main__':
    crawl=Webcrawler()
    crawl.crawler('https://www.cnn.com')
```
